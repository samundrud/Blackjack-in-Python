# Blackjack Game

* This is a simple text-based [BlackJack](https://en.wikipedia.org/wiki/Blackjack) game
* The game has one player versus an automated dealer.


```python
# get deck of cards
import deck_of_cards
from deck_of_cards import deck_of_cards

# get clear output
from IPython.display import clear_output
```

```python
## make object that assigns value (points) to new card

###################################################################

class Card:
    def __init__(self, newcard):
        self.newcard = newcard
        self.name = newcard.name
        self.rank = newcard.value

        if self.rank == 1: # this is the Ace, which has a value of 11 (and can be adjusted to 1, see later)
            self.value = 11
        elif self.rank < 11:
            self.value = self.rank
        else:
            self.value = 10 # these are the face cards (other than Ace)
```

```python
## test Card object
Deck = deck_of_cards.DeckOfCards()
Deck.shuffle_deck()
newcard = Deck.give_first_card()
card1 = Card(newcard)

print(card1.name)
print(card1.rank)
print(card1.value)
```

```python
# function to play game
# used for player (player decides whether to play or not)
# used for dealer (automated)

def play():
    global points
    # pick up a new card from deck
    newcard = Deck.give_first_card()
    print(f"Hit card: {newcard.name}\n")
    
    # add card name to cards and rank to cards_ranks
    cards.append(newcard.name)
    cards_ranks.append(newcard.value)
    
    # add value (i.e., # of points) of new card to cards_values
    if newcard.value == 1: # this is the Ace
        cards_points.append(11)
    elif newcard.value < 11:
        cards_points.append(newcard.value)
    else:
        cards_points.append(10) # these are the face cards (other than Ace)

    # add up the points
    points = sum(cards_points)

    # now deal with the Ace (if there is one) 
    # if the total sum is now > 21, the Ace will count as 1, so I substract 10 points from the total count
    if points > 51 and cards_ranks.count(1) == 4: # four aces
        # adjust points by taking off 40 points (this makes all four Aces count as 1 instead of as 11)
        points -= 40
    elif points > 41 and cards_ranks.count(1) == 3: # three aces
        # adjust points by taking off 30 points (this makes all three Aces count as 1 instead of as 11)
        points -= 30
    elif points > 31 and cards_ranks.count(1) == 2: # two aces
        # adjust points by taking off 20 points (this makes both Aces count as 1 instead of as 11)
        points -= 20
    elif points > 21 and 1 in cards_ranks:
        # adjust points by taking off 10 points (this makes the Ace count as 1 instead of as 11)
        points -= 10
    else:
        pass
```

```python
# make player bank account and keep track of money
class Account:
    
    def __init__(self, balance):
        self.balance = balance
        
    def __str__(self):
        return f'Account balance: ${self.balance}'
        
    def winnings(self, win):
        self.balance = self.balance + win
        #return print(f'You won ${win}!')
    
    def bet(self, bet):
        if bet > self.balance:
            pass
            #return print(f'You only have ${self.balance}, sucker...!')
        else:
            self.balance = self.balance - bet
            return print(f'You bet ${bet}')
        
account_on = False
```

## Play the game

```python
print('>>> Welcome to Blackjack <<<\n\n')
game_on = input('Do you want to play the game? \nEnter Yes or no: ').lower().startswith('y')
replay = True
account_on = False

while replay:
    game_on  = input('Do you want to play the game? \nEnter Yes or no: ').lower().startswith('y')
    if not game_on:
        print(f'You walk away with ${account.balance}!')
        replay = False
        break
    elif account.balance == 0:
        print("Sorry, you have no more money.")
        replay = False
        break
    while game_on:
        clear_output()
        while account_on == False:
            balance = int(input('How much money do you have? '))
            account = Account(balance)
            account_on = True

        print('\n>>> Welcome to Blackjack <<<\n')
        print(account)

        # get a fresh deck and shuffle it
        Deck = deck_of_cards.DeckOfCards()
        Deck.shuffle_deck()

        '''
        Part 1 of game: The dealer gets the first two cards (one shown and one hidden)
        '''
        # Dealer picks up two cards
        card1 = Card(Deck.give_first_card())
        card2 = Card(Deck.give_first_card())

        # make lists that store dealer's cards (1 each for card name, rank, and value)
        cards_D = [card1.name, card2.name]
        cards_D_ranks = [card1.rank, card2.rank]
        cards_D_points = [card1.value, card2.value]

        ## PRINT OUT THE CARDS OF THE DEALER
        print('\n                 DEALER')
        print('  Card 1                          Card 2')
        print(F'{cards_D[0]}               Card is hidden\n')
        #print(f'There are now {len(Deck.deck)} cards remaining in the deck.')


        '''
        Part 2 of game: The player bets money, then gets the first two cards
        '''
        # ask player how much they want to bet
        bet = int(input('How much money do you want to bet? '))
        while bet > account.balance:
            bet = int(input(f'You only have ${account.balance} left in your account. \nHow much money fo you want to bet? '))
        account.bet(bet)

        # player gets two cards
        card1 = Card(Deck.give_first_card())
        card2 = Card(Deck.give_first_card())

        # make lists that store players cards (1 each for card name, rank, and value)
        cards = [card1.name, card2.name]
        cards_ranks = [card1.rank, card2.rank]
        cards_points = [card1.value, card2.value]

        # add up player score so far
        points = sum(cards_points)

        ## PRINT OUT THE CARDS OF THE PLAYER
        print('\n                 PLAYER')
        print('  Card 1                          Card 2')
        print(F'{cards[0]}                 {cards[1]}\n')
        print(f'Total points so far: {points}')
        #print(f'There are now {len(Deck.deck)} cards remaining in the deck.')

        '''
        Part 3 of game: The player plays the game: hit or stand
        '''
        hit = input('Do you want to HIT or STAND? \nEnter H for HIT or any other key for STAND: ').lower().startswith('h')

        while hit == True:
            play()
            print(f'Total points: {points}')
            print(f'Your cards: {cards}\n')
            points_P = points
            if points > 21:
                print(f'BUST! You lost ${bet}. \nThere are ${account.balance} left in your account.')
                game_on = False
                hit = False
                break
            else:
                hit = input('Enter H for HIT or any other key for STAND: ').lower().startswith('h')
                continue

        else:
            print(f'\nYou stand. Total points: {points}')
            print(f'Your cards: {cards}\n')
            points_P = points
            hit = False

        if not game_on:
            break

        else:
            '''
            Part 4: The dealer plays the game according to certain rules - AUTOMATED DEALER
            '''
            # add up the points (DEALER)
            # show second dealer card
            # if the total value of the dealer cards is < 17, the dealer picks up another
            # continues until value >= 17
            # if there is an Ace => soft hand: Ace can count as 1 or 11
            # rules vary, but here, if dealer points are => 17 (and Ace = 11), dealer will stand

            points_D = sum(cards_D_points)
            print('\n\nDEALERs first two cards:')
            print(F'{cards_D[0]}                 {cards_D[1]}\n')

            # prepare for function:
            points = points_D
            cards = cards_D
            cards_ranks = cards_D_ranks
            cards_points = cards_D_points


            # pick up more cards while points_D < 17 (while loop)
            while points < 17:
                play()

            else:
                game_on = False
                print("The dealer picks up no more cards")


            # print out dealers cards and points

            print("\nDEALER's final cards:")
            print(cards)    
            print(f"\nDEALER's total points: {points}")
            print(f"YOUR total points: {points_P}")
            points_D = points

            # print out who wins and add money to account
            if points_D > 21:
                print (f'\nDealer BUSTS\nCONGRATULATIONS, YOU WON ${bet}!')
                account.winnings(2*bet)
                print(account)
            elif points_P > points_D:
                print(f'\nCONGRATULATIONS, YOU WON ${bet}!')
                account.winnings(2*bet)
                print(account)
            elif points_P < points_D:
                print(f'\nSORRY, YOU LOST ${bet}')
                print(account)
            else:
                print("\nIt's a tie...")
                account.winnings(bet)
                print(account)




```
